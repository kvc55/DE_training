```python
import pandas as pd
from urllib.error import HTTPError
```

```python
try:
    medals_df = pd.read_csv('http://winterolympicsmedals.com/medals.csv')
except HTTPError:
    print('Error when trying to download .csv file')
except Exception as e:
    print('Error')
```
